## This appliaction shuold display a dog image...using skimage and matplotlib packegs

if skimage or matplotlib not installed before, uncomment the first 2 lines for instalation

```python
# import sys
# !{sys.executable} -m pip install scikit-image matplotlib

from skimage import io
import matplotlib.pyplot as plt

# Define the URLs of the images
img_urls = [
    'https://raw.githubusercontent.com/morgalut/img_dog/main/img/dog.webp',
    'https://raw.githubusercontent.com/morgalut/img_dog/main/img/dog1.webp',
    'https://raw.githubusercontent.com/morgalut/img_dog/main/img/download.webp'
]

# Create a figure with subplots
fig, axes = plt.subplots(1, len(img_urls), figsize=(15, 5))

# Loop through the images and display them
for i, img_url in enumerate(img_urls):
    img = io.imread(img_url)
    axes[i].imshow(img)
    axes[i].axis('off')  # Turn off axis labels and ticks
    axes[i].set_title(f'Image {i+1}')

plt.show()
```
