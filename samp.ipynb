```python
import skimage.data as skd
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, IntSlider

# Function to adjust colors
def adjust_colors(image, red_change=0, green_change=0, blue_change=0):
    modified_img = np.copy(image)
    modified_img[:,:,0] = np.clip(modified_img[:,:,0] + red_change, 0, 255)  # Red channel
    modified_img[:,:,1] = np.clip(modified_img[:,:,1] + green_change, 0, 255)  # Green channel
    modified_img[:,:,2] = np.clip(modified_img[:,:,2] + blue_change, 0, 255)  # Blue channel
    return modified_img

# Load the astronaut image
img1 = skd.astronaut()

# Create a figure with multiple subplots
plt.figure(figsize=(12, 4))

# Function to display image with color adjustments
def display_with_adjustments(red_change=0, green_change=0, blue_change=0):
    plt.clf()
    modified_img = adjust_colors(img1, red_change, green_change, blue_change)
    plt.imshow(modified_img)
    plt.axis('off')
    plt.show()

# Original Image
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img1)
plt.axis('off')

# Display adjusted image
plt.subplot(1, 2, 2)
plt.title("Adjusted Image")
display_with_adjustments()

# Define sliders for adjusting colors
red_slider = IntSlider(min=-100, max=100, step=1, description='Red:')
green_slider = IntSlider(min=-100, max=100, step=1, description='Green:')
blue_slider = IntSlider(min=-100, max=100, step=1, description='Blue:')

# Connect sliders to the function
interact(display_with_adjustments, red_change=red_slider, green_change=green_slider, blue_change=blue_slider)

plt.show()
```
